## Create the Following Dataframe

```python
# Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as Instrumentation and hometown Luzon

Instru = all[(all.Track == 'Instrumentation') 
            & (all.Hometown == 'Luzon') 
            & (all.Electronics > 70)][['Name', 'GEAS', 'Electronics']].reset_index(drop=True)
Instru

```

```python
# Filename: Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

# Add an 'Average' collumn at dataframe all
all['Average'] = all[['Math', 'Electronics', 'GEAS','Communication']].mean(axis=1)
all
```

```python
# Filter the values to only show what is needed
# [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is constant as Mindanao and gender Female

Mindy = all[(all.Hometown == 'Mindanao')
            & (all.Gender == 'Female')
            & (all.Average >= 55)][['Name', 'Track', 'Electronics','Average']].reset_index(drop=True)
Mindy
```

## Create a Visualization
Create a visualization that shows how the different features contributes to average grade. Does chosen track in college, gender, or hometown contributes to a higher average score?

```python
fig1 = plt.figure(figsize=(20,5))
colors = ['blue', 'yellow', 'orange']
plt.bar(all['Name'], all['Average'], color =  colors)
plt.title('Average Grade of the Different Students')
```

### College Track

```python
# Create a figure with size 20 x 5 (2o width, 5 length)
fig2 = plt.figure(figsize=(20,5))

# Group by track and create a box plot using average scores as content in the Dataframe 'all'
grouped = all.groupby('Track')['Average'].apply(list)

# Prepare data for plotting
plot_data = [group for group in grouped]

# Create box plot
fig2 = plt.figure(figsize=(20, 5))
plt.boxplot(plot_data, labels=grouped.index)
```

Using this boxplot, we notice that on average, those in **Microelectronics** tracks score more that those in **Communication** and **Instrumentation**. Hence by choosing **Microelectronics**, we can infer that there is a unique advantage upon taking the board exams. 

### Gender

```python
# Create a figure with size 20 x 5 (2o width, 5 length)
fig3 = plt.figure(figsize=(20,5))

# Group by track and create a box plot using average scores as content in the Dataframe 'all'
grouped = all.groupby('Gender')['Average'].apply(list)

# Prepare data for plotting
plot_data = [group for group in grouped]

# Create box plot
fig2 = plt.figure(figsize=(20, 5))
plt.boxplot(plot_data, labels=grouped.index)
```

In this box plot, we notice that **Male** gender has a slightly higher 50th percentiles than **Female** gender. However, those that have high averages are dominated by the **Female** gender, rather than **Male** gender. ***Male** is more compact while **Female** is more spreadout. We can infer that gender makes little difference in taking a boards.

### Hometown

```python
# Create a figure with size 20 x 5 (2o width, 5 length)
fig3 = plt.figure(figsize=(20,5))

# Group by track and create a box plot using average scores as content in the Dataframe 'all'
grouped = all.groupby('Hometown')['Average'].apply(list)

# Prepare data for plotting
plot_data = [group for group in grouped]

# Create box plot
fig2 = plt.figure(figsize=(20, 5))
plt.boxplot(plot_data, labels=grouped.index)
```

With the help of the boxplot, we notice that **Luzon** is the most spreadout while **Visayas** and **Mindanao** are more compact. However, **Luzon** takes monopoly of higher averages with a higher 50th percentile and ceiling. We can infer that living in **Luzon** can have a significant advantage, yet is highly subjective.
